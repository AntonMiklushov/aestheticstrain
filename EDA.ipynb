Задача регрессии эстетического качества изображений.
Требуется по входному цветному изображению предсказать действительное число – усреднённую оценку визуальной привлекательности (aesthetic score) по шкале 1–10, основанную на голосах людей из датасета AVA.

Метрики используются стандартные для регрессии:
  - $MSE = \frac1N \sum\limits_{i}(\hat y_i - y_i) ^ 2$
  - $RMSE = \sqrt{MSE}$
  - $MAE = \frac1N \sum\limits_{i} |\hat y_i - y_i|$

Можно было бы еще добавить R^2, но он не так интуитивен в шкале 1-10

В качестве основы для обучения используется ResNet-50. Это сверточная нейросеть с остаточными блоками, широко используемая как базовая архитектура для задач компьютерного зрения. Она используется в предобученном на ImageNet виде, она с самого начала умеет выделять общие визуальные признаки, задача в том, чтобы адаптировать ее к задаче эстетике, выходной слой заменен на один регрессионный нейрон, и дальше обучается на AVA.

PyTorch и torchvision это база (ничего лучше я не придумал, еще есть tesor flow, но все нейронки для изображений, что я запускал типа flux и stable diffusion вроде были на troch, поэтому торч)

Задача интересна тем, что таргет не какой-то математический параметр, а субъективная человеческая оценка.

Сам датасет содержит в себе 30 гигабайтов картинок разного разрешения и содержания и файл с человеческими оценками для каждой картинки (для каких-то оценок больше, для каких-то меньше, это тоже желательно будет учесть во время обучения)
